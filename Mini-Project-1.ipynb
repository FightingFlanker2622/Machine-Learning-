# Function Building

```python
def classify_all_addiction_scores(score_group_df):
    """
    Classifies addiction levels for all students and stores them in a DataFrame called 'risk_level'.

    Parameters:
    - score_group_df (pd.DataFrame): Must have student_id as index and 'Addicted_Score' column.

    Returns:
    - risk_level (pd.DataFrame): DataFrame containing classification results.
    """
    results = []

    for student_id in score_group_df.index:
        addiction_score = score_group_df.loc[student_id, 'Addicted_Score']

        # Classify the addiction level
        if 1 <= addiction_score <= 3:
            level = "Low"
        elif 4 <= addiction_score <= 6:
            level = "Medium"
        else:
            level = "High"

        results.append({
            "Addiction_Level": level
        })

    # Convert results into a DataFrame named 'risk_level'
    risk_level = pd.DataFrame(results)

    return risk_level
risk_level = classify_all_addiction_scores(df[['Addicted_Score']])

# Preview result
print(risk_level.head())


```
